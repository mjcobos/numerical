# Suma de Matrices
Problema: Se requiere conocer el consumo de una empresa ganadera (gastos en la alimentacin de los animales)


```python
a = [[21, 18, 35, 40], [19, 11, 21, 14], [12, 15, 20, 30]]
b = [[11, 7, 21, 32], [9, 15, 24, 10], [23, 8, 12, 22]]

def matrix_sum(m1, m2):
    if len(m1) == len(m2) and len(m1[0]) == len(m2[0]): ##mismas num filas y mismas num columnas##
        m3 = []
        for i in range(len(m1)):
            m3.append([])
            for j in range(len(m1[0])):
                m3[i].append(m1[i][j] + m2[i][j])
        return m3
    else:
        return None

c = matrix_sum(a, b)
if c == None:
    print("error")
else:
    for row in c:
        print("[", end=" ")
        for element in row:
            print(element, end=" ")
        print("]")
```

# Multiplicacin de Matrices


```python
a = [[1,2,3], [4,5,6]]
b = [[1,2], [3,4], [5,6]]

def matrix_product(m1, m2):
    if len(m1[0]) == len(m2): ##mismas num columnas con filas
        m3 = []
        for i in range(len(m1)):
            m3.append([])
            for j in range(len(m2[0])):
                m3[i].append(0)
                
        for i in range(len(m1)):
            for j in range(len(m2[0])):
                for k in range(len(m1[0])):
                   # print((i,j), (i,k), (j,k), (k,j))
                    m3[i][j] += m1[i][k] * m2[k][j]
        return m3
    else:
        return None
        
c = matrix_product(a,b)
        
if c == None:
    print("error")
else:
    for row in c:
        print("[", end=" ")
        for element in row:
            print(element, end=" ")
        print("]")
   

```
